## QUESTION 3 - What is the average of words per line for each character?

```python
# Load the list of stopwords
with open("C:\\Users\\kdarw\\exam\\data-prep-cleaning-vis-llllym\\stopwords.json", 'r') as file:
    stopwords = json.load(file)

# Define a function to count non-stopwords
def count_non_stopwords(line):
    words = line.split()
    non_stopwords = [word for word in words if word.lower() not in stopwords]
    return len(non_stopwords)


# Apply the function to count non-stopwords for each line
office['non_stopword_count'] = office['line_text'].apply(count_non_stopwords)

# Calculate the total number of non-stopwords per speaker
total_non_stopwords_per_speaker = office.groupby('speaker')['non_stopword_count'].sum()

# Calculate the number of lines per speaker (as before)
lines_per_speaker = office.groupby('speaker')['line_text'].count()

# Calculate the average non-stopwords per line for each speaker
average_non_stopwords_per_line = total_non_stopwords_per_speaker / lines_per_speaker

# Sort the average non-stopwords per line in descending order (optional)
sorted_average_non_stopwords = average_non_stopwords_per_line.sort_values(ascending=False)

# Print the sorted averages
print(sorted_average_non_stopwords)
```

## QUESTION 7 - The average percent of lines each character contributed each episode per season.

# Each character contribution per each episode per season in %
1. Total number of lines per each episode per season.
2. Total number of lines spoken by each speaker per episode per season.
3. step2 / step 1 = gives % 

```python

# Grouping the data by season, episode, and speaker, and count the lines each character contributed
lines_per_character_per_episode = office.groupby(['season', 'episode', 'speaker']).size().reset_index(name='lines')
lines_per_character_per_episode

```

```python
# total lines in a episode
total_lines_per_episode = lines_per_character_per_episode.groupby(['season', 'episode'])['lines'].sum().reset_index(name='total_lines')
total_lines_per_episode
```

```python
# Merge the two DataFrames on season and episode
merged_df = pd.merge(lines_per_character_per_episode, total_lines_per_episode, on=['season', 'episode'])
merged_df
```

```python
# Calculate the percentage of lines each character contributed in each episode
merged_df['percentage'] = (merged_df['lines'] / merged_df['total_lines']) * 100

```

```python
# Calculate the average percentage of lines each character contributed per season
average_percentage_per_season = merged_df.groupby(['season', 'speaker'])['percentage'].mean().reset_index()

# Print the average percentage of lines each character contributed each episode per season
print(average_percentage_per_season)
```
